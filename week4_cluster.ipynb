```python
import pandas as pd
```

```python
df = pd.read_csv('movieLens.txt', sep='|', header=None, quotechar='\"')
```

```python
df.head()
```

```python
df.columns = [
    'ID',
    'Title',
    'ReleaseDate',
    'VideoReleaseDate',
    'IMDB',
    'Unknown',
    'Action',
    'Adventure',
    'Animation',
    'Childrens',
    'Comedy',
    'Crime',
    'Documentary',
    'Drama',
    'Fantasy',
    'FilmNoir',
    'Horror',
    'Musical',
    'Mystery',
    'Romance',
    'SciFi',
    'Thriller',
    'War',
    'Western',
]
```

```python
df.head()
```

```python
len(df['Title'].unique())
```

```python
df.shape
```

```python
df.drop_duplicates(subset=['Title'], keep='first', inplace=True)
```

```python
df.head()
```

```python
X = df.drop(['ID', 'Title', 'ReleaseDate', 'VideoReleaseDate', 'IMDB', 'Unknown'], axis=1)
```

```python
X
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# generate the linkage matrix using ward clustering pre-computed distances
# word is a sophisticated version of centroid clustering
Z = linkage(X, method='ward', metric='euclidean')

# plot the dendrogram
plt.figure(figsize=(10, 7))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('movie')
plt.ylabel('distance')
dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```

```python
# cut tree
from scipy.cluster.hierarchy import cut_tree
```

```python
clusters = cut_tree(Z, n_clusters=10).flatten()
```

```python
# create a table to show the genre distribution in each cluster
X['clusters'] = clusters
```

```python
X.head()
```

```python
X.groupby('clusters').mean().transpose()
```

```python
X.groupby('clusters').size()
```
